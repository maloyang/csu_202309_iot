```python
import pandas as pd
```

```python
#url = "https://odws.hccg.gov.tw/001/Upload/25/opendataback/9059/59/05c978ab-e430-43cd-87a1-8e4e867c9fae.csv"
url = 'https://raw.githubusercontent.com/maloyang/csu_202309_iot/main/W01/bike_data.csv'
df = pd.read_csv(url, encoding='big5')
```

```python
import folium
import geocoder
```

```python
# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新竹市').latlng
```

```python
location
```

```python
m = folium.Map(location=location, zoom_start=16)
```

```python
for i, row in df.iterrows():
    try:
        folium.Marker([float(row['緯度']), float(row['經度'])], popup=row['站點名稱']).add_to(m)
    except Exception as e:
        print(e.args)
```
